```python
### Let check the statsitcs of linear model

import statsmodels.api as sn
```

```python
#### by default stats model do not contain beta not constant or intercept, so we defined it. 
x = sn.add_constant(dt1['room_num'])
```

```python
### lets fit the model in ordinary least square (OLS stands for Ordinary Least Squares, and 
### it is a method used for estimating the parameters in a linear regression model. In the context 
### of linear regression, the goal is to find the best-fitting line through the data by minimizing 
### the sum of the squared differences (residuals) between the observed values and the values predicted by the model.)
### x is  independent variables (features)
### dependent variable (response variable) 
### fit(): This method fits the linear regression model to the data.
lm = sn.OLS(dt1['price'], x).fit()
```

```python
lm.summary()
```

```python
### Interprate.
## 1) Model is OLS (Ordinary least square)
## 2) Beta knot (-34.6592) or intercept
## 3) Beta 1 (room_num) (9.0997)(slope)(it means if x=1 so y will increase by beta1 quantity that is 9.0997)
## 4) Standard Error (it measures the error or variability of predicted and actual value in the model, the model is not 
##     completely describe the result accurately, it will differ by 0.418 in beta 1 and 2.642 in intercept, which means model could
##     not completetly capture the variability of dependent variable).
## 5) t is calculated using beta one/ standard error, so high error means low t which is not good so high t would be favorable.
##    t-statistics, P>|t| values: These values assess the statistical significance of each coefficient. A low p-value (typically below 0.05) 
##    suggests that the corresponding variable is statistically significant.
## 6) p is a threshold point which tells if we get p value below than 1% or 5%, that tell the model has fitted well.
## 7) R squared value ( R squared is a measure of how well the independent variables explain the variability of the dependent variable
##    (It ranges from 0 to 1).1 means model understand the variability well whereas 0 means model have not understood well. )
## 8) Adjusted R squared means it same as R squared with a small difference like the (The Adjusted R-squared is a modified version of 
##    the R-squared (coefficient of determination) in the context of linear regression models. While R-squared tells you the proportion of 
##    the variance in the dependent variable explained by the independent variables, Adjusted R-squared takes into account the number of 
##    predictors (independent variables) in the model, providing a more nuanced measure of model performance.) 
##   The concept of Adjusted R squared
##   Adjusted R-squared adjusts the R-squared value to account for the number of predictors in the model. It penalizes 
##   the use of additional predictors that do not contribute significantly to explaining the variance in the dependent variable.
## 9) The F-statistic tests the overall significance of the regression model. A low p-value (close to zero) suggests that at 
##   least one independent variable has a significant effect on the dependent variable.
```
